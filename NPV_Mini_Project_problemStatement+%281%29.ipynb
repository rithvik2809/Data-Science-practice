#### F. If you order grilled chicken rather than crispy chicken, what is the change in calorie intake?

        Step2: Find the average calories for Crispy Chicken Category

        Step3: Display the suitable graph for the number of calories in Grilled Chicken category

        Step4: Find the average calories for Grilled Chicken Category

        Step5: Get the final visual comparison of Crispy vs Grilled on the basis of calories

        Step6: Write your observations.

##### Step1: Display the suitable graph for the number of calories in Crispy Chicken category

```python
crispy_chicken_items = carbo.loc[carbo['Item_in_Menu'].str.contains("Crispy Chicken", case=False)]['Item_in_Menu']
crispy_chicken_calories = carbo.loc[carbo['Item_in_Menu'].str.contains("Crispy Chicken", case=False)]['Calories']
plt.bar(crispy_chicken_items,crispy_chicken_calories,edgecolor='black')
plt.title('Graph for the number of calories in Crispy Chicken category')
plt.xticks(rotation=90)
plt.show()
```

##### Step2: Find the average calories for Crispy Chicken Category

```python
avg_crispyChicken_calories = carbo.loc[carbo['Item_in_Menu'].str.contains("Crispy Chicken", case=False)]['Calories'].mean()

print('The average calories for Crispy Chicken Category is',round(avg_crispyChicken_calories,2))
```

##### Step3: Display the suitable graph for the number of calories in Grilled Chicken category

```python
grilled_chicken_items = carbo.loc[carbo['Item_in_Menu'].str.contains("Grilled Chicken", case=False)]['Item_in_Menu']
grilled_chicken_calories = carbo.loc[carbo['Item_in_Menu'].str.contains("Grilled Chicken", case=False)]['Calories']
plt.bar(grilled_chicken_items,grilled_chicken_calories,edgecolor='black')
plt.title('Graph for the number of calories in Grilled Chicken category')
plt.xticks(rotation=90)
plt.show()
```

##### Step4:Find the average calories for Grilled Chicken Category

```python
avg_grilledChicken_calories = carbo.loc[carbo['Item_in_Menu'].str.contains("Grilled Chicken", case=False)]['Calories'].mean()

print('The average calories for Grilled Chicken Category is',round(avg_grilledChicken_calories,2))
```

##### Step 5: Get the final visual comparison of Crispy vs Grilled on the basis of calories

```python
crispy_chicken_calories = carbo.loc[carbo['Item_in_Menu'].str.contains("Crispy Chicken", case=False)]
grilled_chicken_calories = carbo.loc[carbo['Item_in_Menu'].str.contains("Grilled Chicken", case=False)]

combined = pd.concat([crispy_chicken_calories,grilled_chicken_calories])
combined_items = combined['Item_in_Menu']
combined_calories = combined['Calories']
plt.bar(combined_items,combined_calories)
plt.grid()
plt.xticks(rotation=90)
plt.show()
```
